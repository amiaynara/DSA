```python
arr=['nemo']
def finding_nemo(arr):
    for i in range(len(arr)):
        if arr[i] == 'nemo' :
            print('Found Nemo')

```

```python
%timeit finding_nemo(arr)
```

```python
# O(nlogn)
def log_pair(arr):
    if len(arr) == 1 :
        return
    print(arr)
    for i in range(len(arr)):
        print(f'({arr[i]}, {arr[0]})')
    log_pair(arr[1:])

a=[1,2,3,4,5]
log_pair(a)
```

```python
 # given two arrays with characters as element
# find whether there is a common element b/w them?

A=['a','b','c','d']
B=['e','f','d']
```

```python
# Naive :
# for each element in A look through all in B
def common(A,B):
    for a in A:
        for b in B:
            if a==b :
                return True
                break
    return False
# O(N*M)
```

```python
# better approach is to use a data structure that can check the presence or 
# absence of an element in O(1)
# Python3: set or dictionary
# so I decide to go with sets 
# Note: set creation is using a list is O(N)
def common_efficient(A,B):
    a_set=set(A)
    b_set=set(B)
    if len(a_set.intersection(b_set)) > 0 :
        return True
    return False
# O(N+M)
```

```python
# yet another solution could be using sets in a different way
def comm(A,B):
    a_set=set(A)
    b_set=set(B)
    # you could try removing A's element from B. if that actually happens return True
    # I mean there are a ton of ideas
    for element in a_set:
        if element in b_set: # O(N)
            return True
    return False

comm(A,B)
# Note: sets are not subscriptable but iterable.
```

```python
import sys
x=1
sys.getsizeof(x) # returns the size of the object in bytes
# note that it also has functions, attributes ...
# unlike C++, where only 4bytes for an integer
```

```python
a=[]
s1=sys.getsizeof(a) # empty list => 64 bytes
a.append(1)
s2=sys.getsizeof(a) # has 96 bytes as a
b=[1]
s3=sys.getsizeof(b) # this has 72 bytes
c=['1']
s4=sys.getsizeof(c)
print(s1,s2,s3,s4)
```
