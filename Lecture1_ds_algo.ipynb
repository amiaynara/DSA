```python
 # given two arrays with characters as element
# find whether there is a common element b/w them?

A=['a','b','c','d']
B=['e','f','d']
```

```python
# Naive :
# for each element in A look through all in B
def common(A,B):
    for a in A:
        for b in B:
            if a==b :
                return True
                break
    return False
# O(N*M)
```

```python
# better approach is to use a data structure that can check the presence or 
# absence of an element in O(1)
# Python3: set or dictionary
# so I decide to go with sets 
# Note: set creation is using a list is O(N)
def common_efficient(A,B):
    a_set=set(A)
    b_set=set(B)
    if len(a_set.intersection(b_set)) > 0 :
        return True
    return False
# O(N+M)
```

```python
# yet another solution could be using sets in a different way
def comm(A,B):
    a_set=set(A)
    b_set=set(B)
    # you could try removing A's element from B. if that actually happens return True
    # I mean there are a ton of ideas
    for element in a_set:
        if element in b_set: # O(N)
            return True
    return False

comm(A,B)
# Note: sets are not subscriptable but iterable.
```

```python
class list_():
    
    def __init__(self):
        self.n=0
        self.size=1
        self.A=self._make_array(self.size) # A receives an array created by the function 
    
    def append(self,item):
        if self.n == self.size :
            # double the size
            B=self._make_array(2*self.size)
            self_size=2*self.size
            # copy the items of the first array to the second one\
            for index,items in enumerate(self.A):
                B[index]=items
            self.A=B
            self.size=2*self.size
        # after the above is satisfied add the item to the next index
        self.A[self.n]=item
        self.n+=1
        
    def _make_array(self, capacity):
        return (capacity*ctypes.py_object)() # just returns an array 
    def __len__(self):
        print(self.n)
        return self.n
    
    def __str__(self):
        cont='[ '
        for i in range(self.n): # note you cannot use 'for i in self.A'
            cont+=str(self.A[i])+', '
        content=cont[:-2]+' ]'
        return  content
    
    def _search(self, item):
        for i in range(self.n):
            if item==self.A[i] :
                return i
        return -1
            
    def index(self,item):
        index=self._search(item)
        if index >=0 :
            return index
        else:
            print(f'"{item}" not found in the list') # a primitive error handling
            return ''
```

```python
# reverse a string
def rev_string(string):
    if type(string)!=str :
        print(string ," is Not a string!!")
        return None
    rev_string=""
    for i in range(len(string)):
        rev_string+=string[len(string)-1-i]
    return rev_string
```

```python
a=[0,3,4,31]
b=[4,6,30]
```

```python
def merge(a,b):
    c=[]
    a_ind=0
    b_ind=0
    c_ind=0
    na=len(a)
    nb=len(b)
    flaga=0
    flagb=0
    while c_ind< na+nb :
        if na == a_ind:
            flaga=1
            break
        if nb == b_ind :
            flagb=1
            break
        
        if a[a_ind]<b[b_ind] :
            # append a[] to c[]
            c.append(a[a_ind])
            a_ind+=1
            c_ind+=1
        elif a[a_ind]>b[b_ind] :
            c.append(b[b_ind])
            b_ind+=1
            c_ind+=1
        elif a[a_ind] == b[b_ind] :
            c.append(b[b_ind])
            c.append(a[a_ind])
            a_ind+=1
            b_ind+=1
            c_ind+=2
    if flaga :
        c.extend(b[b_ind:])
    if flagb :
        c.extend(a[a_ind:])
    return c
```

```python
print(merge(a,b))
```

```python
# find the maximum subarray
############ Kaden's Algo ###########
def max_subarray(nums):
    S=[]
    S.append(nums[0])
    n=len(nums)
    for i in range(1,n):
        if nums[i] + S[i-1] <= nums[i] :
            S.append(nums[i])
        else: 
            S.append(nums[i]+S[i-1])
    # now loop through the array to find out the subarray
    m=max(S)
    k=S.index(m) # index of max-ending
    summ=0
    while summ!=m:
        summ+=nums[k]
        k-=1
    print(nums[k+1:S.index(m)+1])
```

```python
# Move zeros
def swap(z_p,nz_p,nums):
    nums[z_p]=nums[z_p]^nums[nz_p]
    nums[nz_p]=nums[z_p]^nums[nz_p]
    nums[z_p]=nums[z_p]^nums[nz_p]
def new_zero_pointer(z_p,nums):
    for i in range(z_p+1,len(nums)):
        if nums[i]== 0 :
            return i
    return len(nums)
def new_nz_pointer(nz_p,nums):
    for i in range(nz_p+1,len(nums)):
        if nums[i]!= 0 :
            return i
    # else no non-zero element found after nz_p
    return -1
def move(nums):


    z_p=nums.index(0)
    for i in range(len(nums)):
        if nums[i]!=0 :
            nz_p=i
            break

    while nz_p!=-1 and z_p!=len(nums):
        if nz_p > z_p :
            swap(z_p,nz_p,nums)
            z_p=new_zero_pointer(z_p,nums)
            nz_p=new_nz_pointer(nz_p, nums)
        else:
            nz_p=new_nz_pointer(nz_p,nums)


    print(nums)
```
